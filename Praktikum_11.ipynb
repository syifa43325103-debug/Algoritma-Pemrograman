Praktikum 01 - Sequential Search pada Data Angka

```python
def sequential_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return f"Data {target} ditemukan di indeks {i}"
    return f"Data  {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(sequential_search(angka, 30))
print(sequential_search(angka, 100))
```

Praktikum 02 - Binary Search pada Data Angka

```python
def binary_search (data, target):
    left, right = 0, len(data)-1
    while left <= right:
        mid = (left + right) // 2

        # Blok if/elif/else harus memiliki indentasi yang sama
        if data[mid] == target:
            return f"Data {target} ditemukan di indeks {mid}"
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(binary_search(angka, 40))
print(binary_search(angka, 100))
```

Praktikum 03 - Sequential Search untuk Data Nama

```python
def sequential_search_nama(data, target):
  for i in range(len(data)):
    if data[i].lower() == target.lower:
      return f"Nama {target} ditemukan di indeks {i}"
  return f"Nama {target} tidak ditemukan"

nama = ["Andi", "Budi", "Cici", "Doni"]
print("Data:", nama)
print(sequential_search_nama(nama, "Cici"))
print(sequential_search_nama(nama, "Eka"))
```

Praktikum 04 - Studi Kasus Nilai Mahasiswa dengan Binary Search

```python
def binary_search_nilai(data,  target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return f"Nilai {target} ditemukan di indeks {mid}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"Nilai {target} tidak ditemukan"

nilai = [60, 70, 75, 80, 85, 90]
print("Data nilai:", nilai)
print(binary_search_nilai(nilai, 85))
print(binary_search_nilai(nilai, 65))
```

Praktikum 06 - Kombinasi Sorting dan Searching

```python
import random

def insertion_sort(data):
  for i in range(1, len(data)):
    key = data[i]
    j = i - 1
    while j >= 0 and data[j] > key:
      data[j + 1] = data[j]
      j -= 1
    data[j + 1] = key
  return data

def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return f"{target} ditemukan di indeks {mid}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"{target} tidak ditemukan"

data = [random.randint(1, 100) for _ in range(15)]
print("Data acak sebelum diurutkan:", data)
urut = insertion_sort(data)
print("Data acak setelah diurutkan:", urut)
target = urut[7]
print(binary_search(urut, target))
```

Variasi Sequential Search

```python
def sequential_search_variasi(data, target):
  comparisons = 0
  for i, x in enumerate(data):
    comparisons += 1
    if x == target:
      return i, comparisons
  return -1, comparisons

data = [10, 20, 30, 40, 50]
target = 30
pos, comparisons = sequential_search_variasi(data, target)
print("Posisi:", pos, "Perbandingan:", comparisons)
```

Variasi Binary Search

```python
def binary_search_variasi(data, target):
  low, high = 0, len(data) - 1
  steps = 0
  while low <= high:
    steps += 1
    mid = (low + high) // 2
    if data[mid] == target:
      return mid, steps
    elif data[mid] < target:
      low = mid + 1
    else:
      high = mid - 1
  return -1, steps

data = [10, 20, 30, 40, 50]
target = 40
pos, steps = binary_search_variasi(data, target)
print("Posisi:", pos, "Langkah:", steps)
```

Studi Kasus Barang Toko

```python
barang = [
    ("Baju", 100000),
    ("Celana", 150000),
    ("Sepatu", 200000),
    ("Topi", 50000),
]

barang.sort(key=lambda x: x[1])

target_harga = int(input("Masukkan harga yang dicari: "))

low, high = 0, len(barang) - 1
posisi = -1
while low <= high:
  mid = (low + high) // 2
  if barang[mid][1] == target_harga:
    posisi = mid
    break
  elif barang[mid][1] < target_harga:
    low = mid + 1
  else:
    high = mid - 1

if posisi != -1:
    nama_barang = barang[posisi][0]
    print(f"Barang dengan harga {target_harga} adalah {nama_barang})")
else:
    print(f"Barang dengan harga {target_harga} tidak ditemukan.")
```

Analisis Kompleksitas

```python
import time
import random

# 1. Fungsi Sequential Search
def sequential_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return i
    return -1

# 2. Fungsi Binary Search
def binary_search(data, target):
    low = 0
    high = len(data) - 1
    while low <= high:
        mid = (low + high) // 2
        if data[mid] == target:
            return mid
        elif data[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# Ukuran data yang akan diuji
ukuran_data = [100, 1000, 10000]

print(f"{'Ukuran Data':<15}  {'Sequential (detik)':<20}  {'Binary (detik)':<20}")

for n in ukuran_data:
    # Generate data dan urutkan (Binary Search wajib urut)
    data_test = sorted(random.sample(range(n * 10), n))

    # Pilih target yang ada di posisi agak akhir agar terlihat perbedaannya
    target = data_test[-1]

    # Hitung waktu Sequential Search
    start_seq = time.perf_counter()
    sequential_search(data_test, target)
    end_seq = time.perf_counter()
    durasi_seq = end_seq - start_seq

    # Hitung waktu Binary Search
    start_bin = time.perf_counter()
    binary_search(data_test, target)
    end_bin = time.perf_counter()
    durasi_bin = end_bin - start_bin

    print(f"{n:<15} | {durasi_seq:<20.8f} | {durasi_bin:<20.8f}")
```

Proyek Mini

```python
def binary_search_mahasiswa(data, target_nilai):
    low = 0
    high = len(data) - 1
    hasil = []

    while low <= high:
        mid = (low + high) // 2
        # Cek apakah nilai di tengah adalah target
        if data[mid][1] == target_nilai:
            # Karena mungkin ada nilai yang sama, kita ambil semua yang cocok
            hasil.append(data[mid])

            # Cek ke kiri dan kanan untuk nilai yang sama (duplikat)
            l = mid - 1
            while l >= 0 and data[l][1] == target_nilai:
                hasil.append(data[l])
                l -= 1
            r = mid + 1
            while r < len(data) and data[r][1] == target_nilai:
                hasil.append(data[r])
                r += 1
            return hasil

        elif data[mid][1] < target_nilai:
            low = mid + 1
        else:
            high = mid - 1
    return hasil

# --- Program Utama ---
daftar_mahasiswa = []

print("=== Input Data Mahasiswa ===")
jumlah = int(input("Masukkan jumlah mahasiswa: "))

for i in range(jumlah):
    print(f"\nData ke-{i+1}")
    nama = input("Nama  : ")
    nilai = int(input("Nilai : "))
    daftar_mahasiswa.append((nama, nilai))

# 1. Urutkan berdasarkan nilai (syarat Binary Search)
# x[1] merujuk pada elemen 'nilai' dalam tuple (nama, nilai)
daftar_mahasiswa.sort(key=lambda x: x[1])

print("\n=== Daftar Mahasiswa (Terurut Berdasarkan Nilai) ===")
for mhs in daftar_mahasiswa:
    print(f"- {mhs[0]}: {mhs[1]}")

# 2. Proses Pencarian
print("\n=== Pencarian Nilai ===")
cari = int(input("Masukkan nilai yang ingin dicari: "))

hasil_cari = binary_search_mahasiswa(daftar_mahasiswa, cari)

print("\n--- Hasil Pencarian ---")
if hasil_cari:
    print(f"Ditemukan {len(hasil_cari)} mahasiswa dengan nilai {cari}:")
    for mhs in hasil_cari:
        print(f"Nama: {mhs[0]}")
else:
    print(f"Mahasiswa dengan nilai {cari} tidak ditemukan.")
```
