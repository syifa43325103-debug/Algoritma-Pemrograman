Praktikum 01 - Bubble Sort

```
def bubble_sort(data):
  n = len(data)
  for i in range (n):
    for j in range(0, n - i - 1):
      if data[j] > data[j + 1]:
        data[j], data [j + 1] = data[j +  1], data [j]
  return data

angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah Bubble Sort:", bubble_sort(angka))
```

Praktikum 02 - Selection Sort

```
def selection_sort(data):
    n = len(data)
    for i in range(n):
      min_index = i
      for j in range(i + 1, n):
        if data[j] < data[min_index]:
          min_index = j
          data[i], data[min_index] = data[min_index], data[i]
    return data

angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

Praktikum 03 - Insertion Sort

```
def insertion_sort(data):
  n = len(data)
  for i in range(1, n):
    key = data[i]
    j = i - 1
    while j >= 0 and data[j] > key:
      data[j + 1] = data[j]
      j -= 1
      data[j + 1] = key
      return data

angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan:", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

Praktikum 04 - Studi Kasus: Mengurutkan Nilai Mahasiswa

```
def bubble_sort_nilai(data):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if data[j][1] < data[j + 1][1]:
        data[j], data[j + 1] = data[j + 1], data[j]
  return data

mahasiswa = [("Andi", 80), ("Budi", 70), ("Cici", 90), ("Doni", 85)]
print("Data sebelum diurutkan:", mahasiswa)
print("Data setelah diurutkan (desc):", bubble_sort_nilai(mahasiswa))
```

Praktikum 06 - Sorting Data Barang (Nama & Harga)

```
def bubble_sort_key(data, key_index=0, reverse=False):
    n = len(data)
    for i in range (n):
      for j in range (0, n - i - 1):
        if reverse:
          if data[j][key_index] < data[j + 1] [key_index]:
            data[j], data[j + 1] = data[j + 1], data[j]
          else:
            if data[j][key_index] > data[j + 1] [key_index]:
              data[j], data[j + 1] = data[j + 1], data [j]
      return data

barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen", 10000)]
print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(), 0))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(), 1, reverse=True))
```

Eksperimen Bubble Sort

```
def bubble_sort_descending(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):

            if data[j] < data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah Bubble Sort Descending:", bubble_sort_descending(angka))
```

Eksperimen Selection Sort

```
 def selection_sort_max_to_end(data):
    n = len(data)

    for i in range(n - 1, 0, -1):
        max_index = 0

        for j in range(1, i + 1):
            if data[j] > data[max_index]:
                max_index = j

        data[i], data[max_index] = data[max_index], data[i]

    return data

angka = [99, 51, 30, 10, 5]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort (Max to End):", selection_sort_max_to_end(angka))
```

Eksperimen Insertion Sort

```
def insertion_sort_string(data):
    n = len(data)

    for i in range(1, n):
        key = data[i]
        j = i - 1

        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1

        data[j + 1] = key

    return data

nama_mahasiswa = ["Cici", "Andi", "Doni", "Budi", "Edo"]
print("Data sebelum diurutkan:", nama_mahasiswa)
sorted_names = insertion_sort_string(nama_mahasiswa)
print("Data setelah Insertion Sort (Alfabetis):", sorted_names)
```

```
import random

def bubble_sort_count(arr):
    perbandingan = 0
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            perbandingan += 1
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return perbandingan

def selection_sort_count(arr):
    perbandingan = 0
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            perbandingan += 1
            if arr[min_idx] > arr[j]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return perbandingan

def insertion_sort_count(arr):
    perbandingan = 0
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1

        while j >= 0:
            perbandingan += 1
            if key < arr[j]:
                arr[j + 1] = arr[j]
                j -= 1
            else:
                break
        arr[j + 1] = key
    return perbandingan

# Pengujian untuk ukuran input 10, 50, dan 100
ukuran_input = [10, 50, 100]

print(f"{'Ukuran':<10} | {'Bubble':<10} | {'Selection':<10} | {'Insertion':<10}")
print("-" * 50)

for n in ukuran_input:
    data_asli = [random.randint(1, 1000) for _ in range(n)]

    b_count = bubble_sort_count(data_asli.copy())
    s_count = selection_sort_count(data_asli.copy())
    i_count = insertion_sort_count(data_asli.copy())

    print(f"{n:<10} | {b_count:<10} | {s_count:<10} | {i_count:<10}")
```

```
daftar_barang = [
    {"nama": "Laptop", "harga": 7500000},
    {"nama": "Mouse Wireless", "harga": 150000},
    {"nama": "Monitor 24 inch", "harga": 2000000},
    {"nama": "Keyboard Mekanik", "harga": 800000},
    {"nama": "Flashdisk 64GB", "harga": 100000},
]

def tampilkan_barang(barang):
    """Menampilkan daftar barang dalam format tabel."""
    print("\n" + "="*40)
    print(f"| {'No.':<4} | {'Nama Barang':<20} | {'Harga':>10} |")
    print("="*40)
    for i, item in enumerate(barang, 1):
        harga_formatted = f"Rp {item['harga']:,.0f}".replace(",", "_").replace(".", ",").replace("_", ".")
        print(f"| {i:<4} | {item['nama']:<20} | {harga_formatted:>10} |")
    print("="*40)

def urutkan_barang(data, kriteria, urutan):
    is_reverse = (urutan.lower() == 'desc')

    if kriteria.lower() == 'nama':
        sorted_list = sorted(data, key=lambda item: item['nama'], reverse=is_reverse)
    elif kriteria.lower() == 'harga':
        sorted_list = sorted(data, key=lambda item: item['harga'], reverse=is_reverse)

    else:
        print("\n[ERROR] Kriteria pengurutan tidak valid. Gunakan 'nama' atau 'harga'.")
        return data
    print(f"\n[INFO] Data berhasil diurutkan berdasarkan {kriteria.upper()} secara {urutan.upper()}.")
    return sorted_list

def main():
    while True:
        print("\n### Sistem Manajemen Barang Sederhana ###")

        tampilkan_barang(daftar_barang)

        print("\nPilihan Pengurutan:")
        print("1. Urutkan berdasarkan NAMA")
        print("2. Urutkan berdasarkan HARGA")
        print("3. Keluar")

        pilihan_utama = input("Pilih kriteria (1/2/3): ")

        if pilihan_utama == '3':
            print("Terima kasih! Program selesai.")
            break

        if pilihan_utama in ['1', '2']:

            kriteria = 'nama' if pilihan_utama == '1' else 'harga'

            print("\nPilihan Urutan:")
            print("A. Ascending (Menaik)")
            print("D. Descending (Menurun)")

            pilihan_urutan = input("Pilih urutan (A/D): ")
            urutan = 'asc' if pilihan_urutan.upper() == 'A' else 'desc'
            data_terurut = urutkan_barang(daftar_barang, kriteria, urutan)

            tampilkan_barang(data_terurut)

        else:
            print("\n[PERINGATAN] Pilihan tidak valid. Silakan coba lagi.")


if __name__ == "__main__":
    main()
```
